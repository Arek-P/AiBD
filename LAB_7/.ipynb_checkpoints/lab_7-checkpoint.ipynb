```python
# Import modułów
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, classification_report
import numpy as np
```

### Zadanie 1

```python
# Wczytanie danych
data = pd.read_csv('SCORES.csv')
df = pd.DataFrame(data)
print(df.head(),'\n')
print(df.describe(),'\n')

# Wstępny wykres danych
df.plot.scatter(x='Hours', y='Scores', grid=True, title='Wstępny wykres danych');
```

```python
# Skrypt
def regresion(X, y, Xlabel, ylabel, Xlim, ylim):
    
    # Podział na zbiory treningowe i testowe
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=0)

    X_train = np.array(X_train).reshape(-1, 1)
    X_test = np.array(X_test).reshape(-1, 1)
    y_train = np.array(y_train).reshape(-1, 1)
    y_test = np.array(y_test).reshape(-1, 1)

    # Tworzenie modelu
    model = LinearRegression().fit(X_train, y_train)

    # Wskaźniki
    print(f'coef -> {model.coef_}')
    print(f'intercept - > {model.intercept_}')

    # Predykcja czesci testowej
    y_pred = model.predict(X_test)

    # porównanie predykcji z posiadanymi danymi
    fig, ax = plt.subplots(figsize=(10,7))
    ax.scatter(X_test, y_test, label='y_test');
    ax.plot(X_test, y_pred, 'g', label='y_pred');
    ax.set_xlim(Xlim)
    ax.set_ylim(ylim)
    ax.set_xlabel(Xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title('Porównanie wartości przewidzianych i rzeczywistych')
    ax.legend()
    ax.grid()

    plt.show()

    # Walidacja predykcji
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    RMSE = np.sqrt(MSE)
    print(f'MAE = {MAE}')
    print(f'MSE = {MSE}')
    print(f'RMSE = {RMSE}')
```

```python
regresion(df.Hours, df.Scores, 'Hours', 'Scores', (0, 10), (0, 100))
```

Coef to współczynniki dla predyktorów do modelu regresji liniowej. Intercept to współczynnik zerowy, miejsce przecięcia z osią y, gdy wszystkie predyktory są równe 0. 

W tym przypadku Scores = coef * Hours + intercept

Wyniki predykcji modelu są bardzo dobre, spowodowane jest to dużą liniowością danych.

```python
# Inne dane
# Wczytanie danych
data = pd.read_csv('SBP.csv')
df = pd.DataFrame(data)
print(df.head(),'\n')
print(df.describe(),'\n')

# Wstępny wykres danych
df.plot.scatter(x='Age', y='SBP', grid=True, title='Wstępny wykres danych');
```

```python
# Regresja
regresion(df.Age, df.SBP, 'Age', 'SBP', (50, 80), (110, 170))
```

W tym przypadku predykcja jest gorsza, danych jest niewiele, a ich ułożenie jest mocno nieliniowe (są "rozrzucone").

### Zadanie 2

```python
# Wczytanie danych
data = pd.read_csv('PETROL.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]]

# Podział na zbiory treningowe i testowe
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=0)

X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

# Tworzenie modelu
model = LinearRegression().fit(X_train, y_train)

# Wskaźniki
print(f'coef -> {model.coef_}')
print(f'intercept - > {model.intercept_} \n')

# Predykcja czesci testowej
y_pred = model.predict(X_test)

# Walidacja predykcji
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
print(f'MAE = {MAE}')
print(f'MSE = {MSE}')
print(f'RMSE = {RMSE}')
```

W tym przypadku predykcja jest słaba. Danych jest niewiele. Dla niektórych, pojedynczych predyktorów dopasowanie było by dobre, ale zebranie ich wszystkich na raz, powoduje spadek jakości dopasowania.

### Zadanie 3

```python
# Wczytanie danych
data = pd.read_csv('HEART.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
# przetworzenie danych
df = df.replace('?', np.nan)
del df['slope']
del df['ca']
del df['thal']
df = df.dropna()
df
```

```python
df = pd.get_dummies(df, columns=['cp', 'restecg'], drop_first=True)
df
```

```python
X = df[list(df.columns[:8]) + list(df.columns[9:])]
y = df[df.columns[8]]

# Podział na zbiory treningowe i testowe
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=0)

X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

# Tworzenie modelu
model = LogisticRegression().fit(X_train, y_train)

# Wskaźniki
print(f'coef -> {model.coef_} \n')
print(f'intercept - > {model.intercept_} \n')

# Predykcja czesci testowej
y_pred = model.predict(X_test)

# Walidacja predykcji
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
print(f'MAE = {MAE}')
print(f'MSE = {MSE}')
print(f'RMSE = {RMSE} \n')
print(classification_report(y_test, y_pred))
```

Błędy nie są duże, predykcja modelu jest dobra.
