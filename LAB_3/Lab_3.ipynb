### 7. Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.

```python
df = pd.read_sql("SELECT country_id FROM country WHERE country IN ('Poland', 'Nigeria', 'Bangladesh')",con=connection)
country_id = tuple(df['country_id'].tolist())

df = pd.read_sql(f"SELECT city_id FROM city WHERE country_id IN {country_id}",con=connection)
city_id = tuple(df['city_id'].tolist())

df = pd.read_sql(f"SELECT address_id FROM address WHERE city_id IN {city_id}",con=connection)
address_id = tuple(df['address_id'].tolist())

df = pd.read_sql(f"SELECT * FROM customer WHERE address_id IN {address_id}",con=connection)
df
```

```python
# Alternatywna metoda
df = pd.read_sql("SELECT customer.* FROM country " +
                 "INNER JOIN city ON country.country_id = city.country_id " +
                 "INNER JOIN address ON city.city_id = address.city_id " +
                 "INNER JOIN customer ON address.address_id = customer.address_id " +
                 "WHERE country.country IN ('Poland', 'Nigeria', 'Bangladesh') " +
                 "ORDER BY customer_id ASC",con=connection)
df
```

Obie metody zwracają te same rezultaty.

### 11. Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

```python
df = pd.read_sql("SELECT country_id FROM country WHERE country = 'United States'",con=connection)
country_id = tuple(df['country_id'].tolist())

df = pd.read_sql(f"SELECT city_id FROM city WHERE country_id = {country_id[0]}",con=connection)
city_id = tuple(df['city_id'].tolist())

df = pd.read_sql(f"SELECT address_id FROM address WHERE city_id IN {city_id}",con=connection)
address_id = tuple(df['address_id'].tolist())

df = pd.read_sql(f"SELECT customer_id FROM customer WHERE address_id IN {address_id}",con=connection)
customer_id = tuple(df['customer_id'].tolist())

df = pd.read_sql(f"SELECT DISTINCT inventory_id FROM rental WHERE customer_id IN {customer_id}",con=connection)
inventory_id = tuple(df['inventory_id'].tolist())

df = pd.read_sql(f"SELECT DISTINCT film_id FROM inventory WHERE inventory_id IN {inventory_id}",con=connection)
film_id = tuple(df['film_id'].tolist())

df = pd.read_sql(f"SELECT DISTINCT category_id FROM film_category WHERE film_id IN {film_id}",con=connection)
category_id = tuple(df['category_id'].tolist())

df = pd.read_sql(f"SELECT name FROM category WHERE category_id IN {category_id}",con=connection)
df
```

```python
# Alternatywna metoda
df = pd.read_sql("SELECT DISTINCT category.name FROM category " +
                 "INNER JOIN film_category ON category.category_id = film_category.category_id " +
                 "INNER JOIN inventory ON film_category.film_id = inventory.film_id " +
                 "INNER JOIN rental ON inventory.inventory_id = rental.inventory_id " +
                 "INNER JOIN customer ON rental.customer_id = customer.customer_id " +
                 "INNER JOIN address ON customer.address_id = address.address_id " +
                 "INNER JOIN city ON address.city_id = city.city_id " +
                 "INNER JOIN country ON city.country_id = country.country_id " +
                 "WHERE country.country = 'United States'"
                 "ORDER BY name ASC",con=connection)
df
```

W klienci z USA wypożyczyli filmy z każdej kategorii. Obie metody dają te same rezultaty.

### 12. Znajdź wszystkie tytuły filmów, w których grał: Olympia Pfeiffer lub Julia Zellweger lub Ellen Presley

```python
df = pd.read_sql(f"SELECT actor_id FROM actor WHERE " +
                 "(first_name = 'Olympia' AND last_name = 'Pfeiffer') OR " +
                 "(first_name = 'Julia' AND last_name = 'Zellweger') OR " +
                 "(first_name = 'Ellen' AND last_name = 'Presley')",con=connection)
actor_id = tuple(df['actor_id'].tolist())

df = pd.read_sql(f"SELECT DISTINCT film_id FROM film_actor WHERE actor_id IN {actor_id}",con=connection)
film_id = tuple(df['film_id'].tolist())

df = pd.read_sql(f"SELECT title FROM film WHERE film_id IN {film_id}",con=connection)
df
```

```python
# Alternatywna metoda
df = pd.read_sql("SELECT film.title FROM film " +
                 "INNER JOIN film_actor ON film.film_id = film_actor.film_id " +
                 "INNER JOIN actor ON film_actor.actor_id = actor.actor_id " +
                 "WHERE " + 
                 "(first_name = 'Olympia' AND last_name = 'Pfeiffer') OR " +
                 "(first_name = 'Julia' AND last_name = 'Zellweger') OR " +
                 "(first_name = 'Ellen' AND last_name = 'Presley')"
                 "ORDER BY title ASC",con=connection)
df
```

Obie metody dały te same rezultaty.
