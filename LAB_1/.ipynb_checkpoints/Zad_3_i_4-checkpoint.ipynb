```python
import matplotlib.pyplot as plt
import numpy as np

# Definicja funkcji
def f(x):
    return x**2 + 5

# Przedziały wykorzystywane do rysowania funkcji
x1 = np.arange(-1, 1, 0.01)
x2 = np.arange(-6, 6, 0.01)
x3 = np.arange(0, 5, 0.01)


# Tworzenie wykresów
axs = plt.figure(figsize=(15, 4)).subplots(1, 3)
axs[0].plot(x1, f(x1))
axs[0].set(xlabel='X', ylabel='Y', title='$f(x)=x^2 + 5$')
axs[0].legend(['f(x)'])
axs[0].grid()

axs[1].plot(x2, f(x2))
axs[1].set(xlabel='X', ylabel='Y', title='$f(x)=x^2 + 5$')
axs[1].legend(['f(x)'])
axs[1].grid()

axs[2].plot(x3, f(x3))
axs[2].set(xlabel='X', ylabel='Y', title='$f(x)=x^2 + 5$')
axs[2].legend(['f(x)'])
axs[2].grid()

plt.show()
```

W zadaniu zdefiniowano funkcję, a następnie wyrysowano ją na trzech różnych przedziałach na osobnych wykresach. Wykresy uzupełniono: tytułem, opisami osi, legendą oraz siatką.

```python
import pandas as pd

# Nazwy kolumn
col_names = ["name", "surname", "age", "sex"]

# Tworzenie pustego dataframe z powyższymi nazwami kolumn
df = pd.DataFrame(columns = col_names)

# Dodawanie 5 rekordów
df.loc[0] = ["John", "Doe", 23, "M"]
df.loc[1] = ["Frank", "McGregor", 10, "M"]
df.loc[2] = ["Carmen", "Sierra", 40, "F"]
df.loc[3] = ["Clara", "Norris", 22, "F"]
df.loc[4] = ["Jack", "Black", 53, "M"]

# Wypisywanie informacji o danych
print('Informacje o danych:')
df.info()
```

```python
print('\nOpis danych:')
df.describe()
```

```python
print('\nTrzy pierwsze rekordy:')
df.head(3)
```

W zadaniu stworzono pusty dataframe z czterema kolumnami. Następnie wprowadzono 5 rekordów. Na koniec wypisano informacje i opis danych oraz 3 pierwsze rekordy.
