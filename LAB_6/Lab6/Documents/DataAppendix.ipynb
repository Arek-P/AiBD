```python
# rozkład wieku klientów
df.Age.plot(kind='density',title='Age', xlim = [18, 76], grid = True)
plt.show()
```

Największą grupę klientów stanowią osoby w wieku około 40 lat. Może to być spowodowane tym, że w tym wieku znaczna część osób założyła rodziny i ma swój dom/mieszkanie, dlatego są one odpowiedzialne, za zakup urządzeń takich jak odkurzacz.

```python
# rozkład ocen
df.Rating.plot(kind='density',title='Rating', xlim = [0, 5], grid = True)
plt.show()
```

Rozkład ocen nie jest symetryczny. Nieznacznie przeważają oceny poniżej średniej z przedziału ocen (2.5). Ma to odwzorowanie w średniej ze wszystkich wystawianych ocen wynoszącej 2.33. Może to świadczyć o tym, iż nabyte produkty nie spełniły oczekiwań większej części klientów.

```python
# rozkład ilości dni od zakupu do oceny
df.Elapsed_Days.plot(kind='density',title='Elapsed_Days', xlim = [1, 16], grid = True)
plt.show()
```

Większość klientów zdecydowała się na ocenę produktu po upływie około 7 dni od zakupu. Spowodowane jest to zapewne tym, iż w tym czasie użyli oni zakupionego produktu kilka razy. Nie każdy używa od razu zakupionego produktu, lub po prostu 3-4 dni nie są wystarczającą ilością czasu pozwalającą, na pełne przetestowanie urządzenia. Niska liczba ocen po więcej niż 10 dniach jest za pewne spowodowana tym, że po tygodniu używania produktu, jest się w stanie go wystarczająco przetestować, a kolejny tydzień nie zmieni już wiele w tym kontekście.

```python
# analiza marek pod względem ocen
brands = df.Brand.unique()
means = []
amount = []

for brand in brands:
    b = df[df.Brand == brand]
    means.append(b.Rating.mean())
    amount.append(b.Rating.count())

# wykresy
width = 0.7
fig, ax = plt.subplots(1, 2, figsize=(16,6))

ax[0].bar(brands, means, width)
ax[0].set_ylabel('Średnia ocena')
ax[0].yaxis.grid(color='gray', linestyle='dashed')
ax[0].set_title('Średnia ocena produktu danej marki')

ax[1].bar(brands, amount, width)
ax[1].set_ylabel('Ilość ocen')
ax[1].yaxis.grid(color='gray', linestyle='dashed')
ax[1].set_title('Liczba ocen produktu danej marki')

plt.show()
```

Średnie oceny produktów poszczególnych marek są do siebie zbliżone. Warto zatem wziąść ilość ocen wchodzących w te średnie. Liczba ocen dla makri Tefal jest największa, zatem średnia ocena ich produktu jest najbardziej wiarygodna. Mimo, że średnia ocena marki Beko  jest największa, to jendak składa się ona z najmniejszej liości ocen (około 3 razy mniej niż dla Tefal). Można zatem wnioskować, że owa średnia ocena nie będzie się plasowała tak wysoko, dla większej próbki danych.

Warto zauważyć, że ponieważ dane o ocenach nie są wybrakowane, liczba ocen produktu danej marki jest zarówno ilością zakupionych produktów tej marki. Można sformułować kolejny wniosek: mimo, że produkty Beko mają najlepszą średnią ocene, to klienci najrzadziej (około 2-3 razy rzadziej od pozostałych) decydują się na zakup produktów Beko. Możliwe, że klienci preferują inne marki, mimo iż jak można wnioskować po pierwszym wykresie produkty marki Beko nie ustępują jakością produktom posostałych producentów. 

```python
# analiza liczby klientów pod względem płci
K = df.Sex[df.Sex == 'K'].count()
M = df.Sex[df.Sex == 'M'].count()
N = df.Sex.isnull().sum()

# wykres
width = 0.6
fig, ax = plt.subplots(figsize=(8,6))

ax.bar(['Kobieta', 'Mężczyzna', 'Brak danych'], [K, M, N], width)
ax.set_ylabel('Ilość')
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.set_title('Liczba klientów z podziałem na płeć', fontsize=16, pad=20)

plt.show()
```

Pomimo pewnego braku danych, można jednoznacznie stwierdzić, że kobiety stanową większość klientów. Może to być spowodowane tym, że kobiety częściej zajmują się obowiązkami domowymi, taimi jak sprzątanie, dlatego są w stanie lepiej dobrać odkurzacz, który będzie dostosowany do ich potrzeb.
