```python
import pandas as pd
import numpy as np
# wczytanie oryginalnych danych
df=pd.read_csv("../OriginalData/15_ZACHODNIOPOMORSKIE.csv")
df
```

```python
# usunięcie kolumny Unnamed: 0, która nic nie wnosi do naszego zestawu danych, są to indeksy, zbędne do dalszej pracy z danymi
del df["Unnamed: 0"]

# zmiana nazw kolumn
df = df.rename({'Dni od zakupu': 'Elapsed_Days', 'Marka': 'Brand', 'Wiek kupującego': 'Age', 'Płeć kupującego': 'Sex', 'Ocena': 'Rating'}, axis=1)
df
```

```python
# zmniana wartosci "bd." na NaN w kolumnie Sex
df.Sex = df.Sex.map({'bd.': np.nan, 'M':'M', 'K':'K'})
df
```

```python
# zapisanie uporządkowanych danych do pliku csv
df.to_csv("../AnalysisData/final_data.csv", index = False)
```
